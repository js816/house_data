```python
# Loading the dependencies
import pandas as pd
```

```python
# Loading in the King Count housing data as a DataFrame
# Data and details can be found here:  https://www.kaggle.com/harlfoxem/housesalesprediction
house_df = pd.read_csv('Resources/kc_house_data.csv')
house_df
```

```python
# Looking at the data types
house_df.dtypes
```

```python
# Dropping some unneeded columns
house_df = house_df.drop(columns=['id', 'date'], axis=1)
house_df
```

```python
# Looking at the summary for the numeric columns
house_df.describe()
```

```python
# Looking at the values for bedrooms
house_df['bedrooms'].value_counts()
```

```python
# Looking at the values for floors
house_df['floors'].value_counts()
```

```python
# Looking at the values for waterfront
house_df['waterfront'].value_counts()
```

```python
# Looking at the values for view
house_df['view'].value_counts()
```

```python
# Looking at the values for condition
house_df['condition'].value_counts()
```

```python
# Looking at the values for grade
house_df['grade'].value_counts()
```

```python
# Looking at the values for yr_built
house_df['yr_built'].value_counts()
```

```python
# Creating new column for price(k)
house_df['price(k)'] = round(house_df['price']/1000, 1)
house_df
```

```python
# Creating new column for sqft_lot(k)
house_df['sqft_lot(k)'] = round(house_df['sqft_lot']/1000, 1)
house_df
```

```python
# Creating new column for sqft_living(k)
house_df['sqft_living(k)'] = round(house_df['sqft_living']/1000, 1)
house_df
```

```python
# Looking at the summary for the numeric columns
house_df.describe()
```

```python
# Dropping additional features not needed for this analysis
house_df = house_df.drop(columns=['price', 'sqft_living', 'sqft_lot', 'sqft_living15', 'sqft_lot15'])
house_df
```

```python
# Exporting the cleaned DataFrame as a CSV file
house_df.to_csv('Resources/kc_house_data_clean.csv', index=False)
```
